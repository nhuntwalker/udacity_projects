<strong>Question 1: What is our independent variable? What is our dependent variable?</strong>

- Our <strong>independent variable</strong> is the color of each word.
- The <strong>dependent variable</strong> is then the amount of time taken to name the colors in the list.

<strong>Question 2</strong>: What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

- An appropriate null hypothesis $H_0$ is that on average, there is no significant increase in reading time between  congruent and incongruent lists. Numerically, this would be $\mu_C \ge \mu_I$.
- An alternate hypothesis $H$ is then that on average the incongruent lists take significantly longer to read than the congruent lists. Numerically, this would be $\mu_C < \mu_I$.
- From information about the initial study, as well as trying the experiment myself, it seems that the incongruent list takes longer for the brain to process. So, to test this, I should assume that there is no significant time lag, and accept significant ($\alpha < 0.05$) negative deviations (congruent time < incongruent time) from $H_0$ as a rejection of $H_0$.
- To test $H$ I intend to do a <strong>left-tailed t-test for 2 dependent samples</strong>. I'm choosing this test instead of the independent samples t-test since the same person is being tested twice in a row on two slightly different but ultimately similar stimuli (repeated measures design). 
- I'm choosing the t-test in lieu of a z-test because these don't represent full populations, the sample size is small (< 30), and we don't have access to descriptive statistics of the full populations.
- I'm also choosing the t-test instead of other types of tests for non-gaussian distributions because there's nothing in the data (or in the hypothesis) to suggest that these data are anything but normally distributed.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
data = pd.read_csv("stroopdata.csv", sep=",")
```

<strong>Question 3</strong>: Report some descriptive statistics regarding this dataset. Include at least one measure of central tendency and at least one measure of variability.

```python
mu = [data.Congruent.mean(), data.Incongruent.mean()]
median = [data.Congruent.median(), data.Incongruent.median()]
std = [np.std(data.Congruent, ddof=1), np.std(data.Incongruent, ddof=1)]

print "Mean Reading Speed-- \n\tCongruent Words: %.3fs | Incongruent Words: %.3fs\n" % (mu[0], mu[1])
print "Median Reading Speed-- \n\tCongruent Words: %.3fs | Incongruent Words: %.3fs\n" % (median[0], median[1])
print "Sample Standard Deviation in Reading Speed-- \n\tCongruent Words: %.3fs | Incongruent Words: %.3fs\n" % (std[0], std[1])
```

<strong>Question 4</strong>: Provide one or two visualizations that show the distribution of the sample data. Write one or two sentences noting what you observe about the plot or plots.

```python
cut = data.Incongruent < 30

fig = plt.figure(figsize=(12,5))
ax = plt.subplot(121)
ax.plot(data.Congruent, data.Incongruent, 'ok', markerfacecolor="None")

m1, b1 = np.polyfit(data.Congruent, data.Incongruent, deg=1)
m2, b2 = np.polyfit(data.Congruent[cut], data.Incongruent[cut], deg=1)
xtest = np.linspace(0, 30, 5)
ytest1 = m1*xtest + b1
ytest2 = m2*xtest + b2
yone = xtest
ax.plot(xtest, ytest1, linestyle="--", color="#FF0000", label="y = %.2fx + %.2f" % (m1, b1))
ax.plot(xtest, ytest2, linestyle=":", linewidth=2, color="g", label="y = %.2fx + %.2f" % (m2, b2))
ax.plot(xtest, yone, color="k", label="y = x")
ax.text(0.9, 0.05, "a", transform=ax.transAxes, fontsize="14", fontweight="bold")

ax.legend(loc="best", fontsize=10)
ax.set_xlabel("Congruent Reading Time (seconds)")
ax.set_ylabel("Incongruent Reading Time (seconds)")
ax.set_xlim(0,26)
ax.set_ylim(11,37)
ax.minorticks_on()

ax = plt.subplot(122)
ax.hist(data.Congruent, bins=np.arange(4, 38, 2), histtype="step", linewidth=2, label="Congruent Sample")
ax.hist(data.Incongruent, bins=np.arange(4, 38, 2), histtype="step", color="r", linewidth=2, label="Incongruent Sample")
ax.plot([mu[0], mu[0]], [0, 7], linestyle="--", linewidth=2, color="b", label="Congruent Mean:\n%.3f$\pm$%.3f s" % (mu[0], std[0]))
ax.plot([mu[1], mu[1]], [0, 7], linestyle="--", linewidth=2, color="r", label="Incongruent Mean:\n%.3f$\pm$%.3f s" % (mu[1], std[1]))
ax.text(0.9, 0.05, "b", transform=ax.transAxes, fontsize="14", fontweight="bold")

ax.legend(loc="best", fontsize=10)
ax.set_xlabel("Reading Time (seconds)")
ax.set_ylabel("Number of People")
ax.minorticks_on()
ax.set_xlim(4, 46)
ax.set_ylim(0, 7)

plt.show()
```

### Plot "a"
- Generally for each participant, a longer reading speed for the congruent word set corresponds to a longer reading speed for the incongruent word set.
- The above point is illustrated by a linear fit to the data, where the slope is positive.
- All incongruent reading speeds are higher than congruent reading speeds, shown by every point being above the one-to-one black solid line.
- The red dashed fit line is heavily affected by two outliers, with incongruent reading speeds that are really long compared to their corresponding congruent reading speeds. The fit without those two outliers is the green dotted line.

### Plot "b"
- The two samples are vaguely normally-distributed. The incongruent sample moreso than the congruent sample.
- The incongruent sample has a wider spread as compared to the congruent sample. This spread is mostly due to the two outliers. The standard deviation from the mean without them is 2.81 seconds.


<strong>Question 5</strong>: Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

```python
def t_statistics(df, alpha=0.05):
    mu1 = df.Congruent.mean()
    n1 = len(df.Congruent)
    
    mu2 = df.Incongruent.mean()
    n2 = len(df.Incongruent)

    d = df.Congruent - df.Incongruent
    s_d = np.std(d, ddof=1)
    
    se = s_d / np.sqrt(len(d))
    t = (mu1 - mu2)/se
    
    pvals = np.array([0.25, 0.2, .15, .1, .05, .025, .02, .01, .005, .0025, .001, .0005])
    tcrits = -np.array([0.685, 0.858, 1.060, 1.319, 1.714, 2.069, 2.177, 2.500, 2.807, 3.104, 3.485, 3.768])
    
    return t, pvals[tcrits > t][-1]

def confidence_interval(indata, CI=0.95):
    mu = indata.mean()
    std_err = np.std(indata, ddof=1)/np.sqrt(len(indata))
    
    pvals = np.array([0.25, 0.2, .15, .1, .05, .025, .02, .01, .005, .0025, .001, .0005])
    conf_ints = np.array([0.5, 0.6, 0.7, 0.8, 0.9, 0.95, 0.96, 0.98, 0.99, 0.995, 0.998, 0.999])
    
    tcrits = np.array([0.685, 0.858, 1.060, 1.319, 1.714, 2.069, 2.177, 2.500, 2.807, 3.104, 3.485, 3.768])
    tcrit = tcrits[conf_ints == CI]
    
    min_t = float(mu - tcrit * std_err)
    max_t = float(mu + tcrit * std_err)
    
    return min_t, max_t
```

```python
t_statistics(data)
```

```python
print confidence_interval(data.Congruent)
print confidence_interval(data.Congruent, CI=0.99)
print confidence_interval(data.Congruent, CI=0.995)
print confidence_interval(data.Incongruent)
print confidence_interval(data.Incongruent, CI=0.99)
print confidence_interval(data.Incongruent, CI=0.995)
```

<strong>My calculated $t$-statistic is roughly -8.021</strong>. For my criteria of $\alpha <$ 0.05, the critical value was -1.714. In fact, my $t$-statistic is far below even the lowest $\alpha$ of 0.0005 ($t_\text{crit}$ of -3.768).

With this in mind, I'm definitely ready to reject the null hypothesis. There is definitely a longer reading time for the incongruent data set. We can see this if we look at the confidence intervals for reading times:
