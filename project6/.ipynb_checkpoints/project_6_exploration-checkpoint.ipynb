```python
pisa_data = pd.read_csv("pisa2012_test.csv")
```

```python
b = list(pisa_data.iloc[0])
```

```python
data_types = []
for ii in range(1, len(b)):
    if type(b[ii]) == np.int64:
        data_types.append("INT")
    
    elif type(b[ii]) == str:
        data_types.append("VARCHAR(255)")
        
    elif (type(b[ii]) == np.float64) or (type(b[ii] == float)):
        data_types.append("DOUBLE")
```

```python
print data_types[:10]
```

```python
input_data_str = ""
for ii in range(len(a)):
    if a[ii] == "REPEAT":
        add_str = "REPEAT_COL" +  " " + str(data_types[ii]) + ", "
    else:
        add_str = str(a[ii]) + " " + str(data_types[ii]) + ", "
    input_data_str += add_str
print input_data_str
```

```python
columns_str = ""
for ii in range(len(a)):
    if a[ii] == "REPEAT":
        columns_str += "REPEAT_COL, "
    else:
        columns_str += str(a[ii]) + ", "
print columns_str
```
