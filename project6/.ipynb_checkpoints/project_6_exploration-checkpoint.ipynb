```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pisa_columns = pd.read_csv("pisadict2012.csv")
```

```python
a = list(pisa_columns[pisa_columns.columns[0]])
```

```python
pisa_data = pd.read_csv("pisa2012_test.csv")
```

```python
b = list(pisa_data.iloc[0])
```

```python
data_types = []
for ii in range(len(b)):
    if type(b[ii]) == "numpy.int64":
        data_types.append("INT")
    
    if type(b[ii]) == "str":
        data_types.append("VARCHAR(255)")
        
    if (type(b[ii]) == "numpy.float64") or (type(b[ii] == "float")):
        data_types.append("DOUBLE")
```

```python
input_data_str = ""
for ii in range(len(a)):
    add_str = str(a[ii]) + " " + str(data_types[ii]) + ", "
    input_data_str += add_str
```
